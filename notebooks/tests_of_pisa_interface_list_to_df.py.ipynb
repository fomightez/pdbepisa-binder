## Test reading from a local file

Get the text you'd get if you got PISA Interface List by **hand copying** it from 'Interfaces' page when analysing that PDB entry at https://www.ebi.ac.uk/msd-srv/prot_int/cgi-bin/piserver , which you get to by pressing `Launch PDBePISA` at https://www.ebi.ac.uk/pdbe/pisa/pistart.html .

Example of the copied text for `6nt7` is assigned to the variable `s` between the triple ticks below.  
Then that copied text is saved in this working directory as `6nt7_interface_list.txt`.

```python
s=''' ## 	 Structure 1 	 × 	 Structure 2 	 interface 
 area, Å2 	 ΔiG 
 kcal/mol 	 ΔiG 
 P-value 	 NHB 	 NSB 	 NDS 	 CSS 
 NN 	 «» 	 Range 	 iNat 	 iNres 	 Surface Å2 	 Range 	 iNat 	 iNres 	 Surface Å2 
 1 		A	 481 	 130 	 21484 	 ◊ 	B	 486 	 131 	 21467 	  4864.7 	  -72.5 	 0.239 	 38 	 14 	 0 	 0.000 
 2 		A	 52 	 15 	 21484 	 ◊ 	[1SY]A:401	 34 	 1 	 712 	  345.1 	  -2.0 	 0.648 	 4 	 0 	 0 	 0.000 
 3 		B	 46 	 15 	 21467 	 ◊ 	[1SY]A:401	 33 	 1 	 712 	  339.9 	  -0.2 	 0.759 	 4 	 0 	 0 	 0.000'''
%store s >"6nt7_interface_list.txt"
```

```python
%run pisa_interface_list_to_df.py 6nt7
```
